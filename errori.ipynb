GESTIONE DEGLI ERRORI

```python
try:
    numero=int(input("Inserisci un numero: "))
    risultato=100/numero
    print(f"Risultato: {risultato}")
except:
    print("Errore input non valido")
```

```python
try:
    user_id=int(input("Inserisci il tuo ID: "))
    print(f"Benvenuto {user_id}")
except ValueError:
    print("L'ID dev'essere un numero intero.")
    print("Esempio: 12838933")
```

```python
try:
    numero=int(input("Inserisci un numero: "))
    risultato=100/numero
    print(f"100/{numero}={risultato}")
except ValueError:
    print("Errore devi inserire un numero valido \nHai inserito testo o caratteri speciali")
except ZeroDivisionError:
    print("Errore, non puoi dividere per 0")
```

```python
try:
    file=open("file.txt", "r")
    contenuto=file.read()
    file.close()
except FileNotFoundError,PermissionError:
    print("Impossibile accedere al file")
    print("Il file non esiste o non hai i permessi")
```

REGISTRIAMO L'ERRORE

```python
try:
    prezzo=float(input("Inserisci prezzo: "))
    quantita=int(input("Inserisci quantità: "))
    totale=prezzo*quantita
    print(f"Totale: {totale}")
except ValueError as e:
    print("Errore di conversione")
    print(f"Dettagli tecnici {e}")
    print(f"Tipo di errore: {type(e).__name__}")
```

```python
print("Registrazione età")
print('_'*30)
try:
    eta=int(input("Inserisci età: "))
except ValueError:
    print("Input non valido")
else:
    print("Età registrata con successo")
    print(f"Hai {eta}")
if eta<18:
    print("Sei minorenne")
elif eta<0 or eta>120:
    print("Hai inserito un età non valida")
else:
    print("Sei maggiorenne")
```
